```python
#Reading the Foreign Direct Investment Dataset
annual_inflation_file= "../Resources/Inflation_annual.csv"
annual_inflation_df = pd.read_csv( annual_inflation_file)
annual_inflation_df.head()
```

```python
foreign_dir_inv_df=foreign_dir_inv_df[["Foreign direct investment, net inflows (% of GDP)","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010"]]
foreign_dir_inv_df=foreign_dir_inv_df.dropna()
foreign_dir_inv_df.rename(columns = {'Foreign direct investment, net inflows (% of GDP)':'Country','2001':'FDI_2001','2002':'FDI_2002','2003':'FDI_2003','2004':'FDI_2004','2005':'FDI_2005','2006':'FDI_2006','2007':'FDI_2007','2008':'FDI_2008','2009':'FDI_2009','2010':'FDI_2010',}, inplace = True)
foreign_dir_inv_df
```

```python
annual_inflation_df=annual_inflation_df[["Inflation, GDP deflator (annual %)","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010"]]
annual_inflation_df=annual_inflation_df.dropna()
annual_inflation_df.rename(columns = {'Inflation, GDP deflator (annual %)':'Country','2001':'Inflation_2001','2002':'Inflation_2002','2003':'Inflation_2003','2004':'Inflation_2004','2005':'Inflation_2005','2006':'Inflation_2006','2007':'Inflation_2007','2008':'Inflation_2008','2009':'Inflation_2009','2010':'Inflation_2010'}, inplace = True)
annual_inflation_df
```

```python
merged_df=pd.merge(foreign_dir_inv_df, annual_inflation_df, on = "Country", how = "inner")
merged_df
```
